```python
import matplotlib.pyplot as plt # import pyplot sub-module from matplotlib module
import numpy as np # import numpy

# Generate some data
mu = 0
sigma = 1
x = np.linspace(-5, 5, 100)
y = np.exp(-(x - mu)**2 / (2 * sigma**2))

plt.plot(x, y) # Plot the data

# Add a title and labels
plt.title("Gaussian Curve") # Title label
plt.xlabel("x") # x-axis label
plt.ylabel("y") # y-axis label
plt.show() # Show the plot
```

```python
from scipy.stats import skewnorm # import skeworm from sub-module stats in module scipy
import matplotlib.pyplot as plt # import pyplot module from matplotlib

numValues = 10000
maxValue = 100
skewness = -5   #Negative values are left skewed, positive values are right skewed.

random = skewnorm.rvs(a = skewness,loc=maxValue, size=numValues)  #Skewnorm function

random = random - min(random)      #Shift the set so the minimum value is equal to zero.
random = random / max(random)      #Standadize all the vlues between 0 and 1.
random = random * maxValue         #Multiply the standardized values by the maximum value.

#Plot histogram to check skewness
plt.hist(random,30,density=True, color = 'red', alpha=0.1)
plt.show()

from scipy.stats import skewnorm # import skeworm from sub-module stats in module scipy
import matplotlib.pyplot as plt # import pyplot module from matplotlib

numValues = 10000
maxValue = 100
skewness = 5   #Negative values are left skewed, positive values are right skewed.

random = skewnorm.rvs(a = skewness,loc=maxValue, size=numValues)  #Skewnorm function

random = random - min(random)      #Shift the set so the minimum value is equal to zero.
random = random / max(random)      #Standadize all the vlues between 0 and 1.
random = random * maxValue         #Multiply the standardized values by the maximum value.

#Plot histogram to check skewness
plt.hist(random,30,density=True, color = 'blue', alpha=0.1)
plt.show()
```

T-Test

A t-test is a statistical calculation that compares the means of two sample groups. It's used to determine if there is a significant difference between the means of two groups and how they are related.

```python
from scipy.stats import ttest_ind # from scipy package module stats impot T test module

group1 = np.random.normal(0, 1, 100) # generate random data
group2 = np.random.normal(1, 1, 100) # generate random data

t_stat, p_value = ttest_ind(group1, group2) # run T Test
print("T-Statistic:", t_stat) # print T-Statistic
print("P-Value:", p_value) # print p-value
```

Paired T-Test

A paired t-test is a statistical procedure that compares the means of two sets of observations. The test is used to determine whether the mean difference between the two sets is zero.

```python
from scipy.stats import ttest_rel # from scipy package module stats import paired T test module

before = np.random.normal(10, 2, 50) # generate random data
after = before + np.random.normal(2, 1, 50) # generate random data

t_stat, p_value = ttest_rel(before, after) #run paired T Test
print("T-Statistic:", t_stat) # print T-Statistic
print("P-Value:", p_value) # print p-value
```

Chi-Square Test

A chi-square test is a statistical test that compares observed results with expected results. The goal of the test is to determine if a difference between observed data and expected data is due to chance or to a relationship between the variables being studied

```python
from scipy.stats import chi2_contingency

observed = np.array([[10, 15], [5, 20]])
chi2, p, dof, expected = chi2_contingency(observed)
print("Chi-Square:", chi2)
print("P-Value:", p)
```

ANOVA

ANOVA stands for Analysis of Variance. It is a statistical method that compares the means of two or more groups. ANOVA is often used to determine if there are statistically significant differences between the means of different groups.

```python
from scipy.stats import f_oneway

group1 = np.random.normal(0, 1, 100)
group2 = np.random.normal(1, 1, 100)
group3 = np.random.normal(2, 1, 100)

f_stat, p_value = f_oneway(group1, group2, group3)
print("F-Statistic:", f_stat)
print("P-Value:", p_value)
```
