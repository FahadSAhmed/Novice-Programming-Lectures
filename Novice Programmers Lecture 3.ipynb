Comparison Operators

```python
x = 10
y = 5

x == y
print(x)

print("x is greater then y:",x > y) #greater than
print("x is equal to y:", x == y) #equal
print("x is not equal to y:", x != y) #not equal
print("x is less then y:",x < y) #less than
print("x is greater or equal to y:", x >= y) #greater than or equal to
print("x is less then or equal to y:", x <= y) #less than or equal to
```

While Loops

```python
count = 0 #variable

while count < 6: #while loop
    print(count) #print statement
    count = count + 1 # increment, other short way of writing is count += 1
```

```python
def greet(name):
    return "Hello, " + name + "!" #function

result = greet("Bob") #function call
print(result) #print result
```

```python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def complex_math(x, y, z, e):
  return x + y + z + e

x = input("X: ") #input function
y = input("Y: ") #input function
z = input("Z: ") #input function
e = input("E: ") #input function

result = complex_math(int(x), int(y), int(z), int(e)) #function call
print("Complex math", result) #print result

result = add(int(x),int(y)) #function call
print(result) #print result

result = subtract(int(x),int(y)) #function call
print(result) #print result
```

```python
def adulthood(name, age):
    if age >= 18:
        return "You are an adult, " + name
    else:
        return "You are a minor, " + name

name = input("Enter your name: ") #input function
age = int(input("Enter your age: ")) #input function

print(adulthood(name, age)) #function call
```

Classes and Inheritance

```python
# prompt: define classes and inheritence

class Animal: #class
    def speak(self):
        pass

class Dog(Animal): #class inheritance
    def speak(self):
        print("Woof!")
        print(1+1)

my_dog = Dog() #object
my_dog.speak() #print woof!

```

Modules and Packages

```python
import math #import math
x = 25 #variable
y = 16 #variable
z = 4 #variable

print(math.sqrt(x)) #print square root of x
print(math.sqrt(y)) #print square root of y
print(math.sqrt(z)) #print square root of z + 1

print(math.log10(x)) #print log10 of x
print(math.log10(y)) #print log10 of y
print(math.log10(z)) #print log10 of z + 1
```
