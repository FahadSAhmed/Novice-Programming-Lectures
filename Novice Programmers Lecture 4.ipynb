Opening a File

```python
# Open a file in read mode
file = open("example.txt", "r") # open the file in read mode
```

```python
print(file) # print the file structure
```

Reading File Content

```python
content = file.read() # read the entire file
print(content)        # print the entire file
file.close()          # close the file
```

Reading Lines from a File

```python
file = open("example.txt", "r") # open the file in read mode
for line in file:               # read the file line by line
    print(line)                 # print the line
file.close()                    # close the file
```

Writing to a File

```python
file = open("output.txt", "w")      # open the file in write mode
file.write("Hello, File Handling! Output!") # write the string "Hello, File Handling!" to the file
file.close()                        # close the file
```

Appending (putting chnages in an exsisting file) to a File

```python
file = open("output.txt", "a")  # open the file in append mode (r = read only, w = write, a = append)
file.write("\nAppended text.")  # write the string "Appended text." to the file, for loop you can append large amount of data in one file.
file.close()                    # close the file
```

```python
with open("output.txt", "r") as file:  # open the file in read mode
    content = file.read()               # read the file
    print(content)                      # print the entire file
```

Using 'with' Statement (Context Managers)

```python
with open("example.txt", "r") as file: # open the file in read mode
    content = file.read()              # read the file
    print(content)                     # print the entire file
```

Seeking and Changing File Position

```python
file = open("example.txt", "r") # open the file in read mode
file.seek(20)                   # Move to the 10th byte
content = file.read()           # read the file
print(content)                  # print the entire file
file.close()                    # close the file
```

Reading and Writing CSV Files (or tab files)

```python
# a csv is a file in which each value is seperated by a ',' comma, comma-delimited
# a tab is a file in which each value is seperated by a ' ' tab, tab-delimited

import csv                              # import the csv module

with open("data.csv", "r") as csv_file: # open the file in read mode
    reader = csv.reader(csv_file)       # read the file
    for row in reader:                  # read the file
        print(row)                      # print the entire file
```

```python
with open("output.csv", "w") as csv_file: # open the file in write mode
    writer = csv.writer(csv_file)         # write the file
    writer.writerow(["Name", "Age"])      # write the first row
    writer.writerow(["Alice", 25])        # write the second row
    writer.writerow(["Bob", 30])          # write the third row
    writer.writerow(["Ali", 25])         # write the fourth row
    writer.writerow(["Dawood", 25])         # write the fifth row
    writer.writerow(["Ibrahim", 25])          # write the sixth row
```

```python
with open("output.csv", "r") as csv_file: # open the file in read mode
    reader = csv.reader(csv_file)       # read the file
    for row in reader:                  # read the file
        print(row)                      # print the entire file
```

Using 'json' Module for JSON Files

```python
import json # import the json module

data = {              # create a dictionary
    "name": "Alice",  # add a key-value pair
    "age": 30         # add a key-value pair
}

with open("data.json", "w") as json_file: # open the file in write mode
    json.dump(data, json_file)            # write the dictionary to the file
```

```python
loaded_data = {}                         # create an empty dictionary, variable

with open("data.json", "r") as json_file: # open the file in read mode
    loaded_data = json.load(json_file)    # load the dictionary

print(loaded_data)                        # print the dictionary
```

Reading and Writing Text Files Line by Line

```python
with open("poem.txt", "w") as file: # open the file in write mode
    file.write("roses are red ...") # write the string to the file
    file.write("\nviolets are blue ...") # write the string to the file
    file.write("\nthe light is bright  ...") # write the string to the file
    file.write("\nand so are you ...") # write the string to the file
```

```python
with open("poem.txt", "r") as file: # open the file in read mode
    lines = file.readlines()        # read the file line by line
    for line in lines:              # read the file line by line
        print(line)                 # print the entire file
```

Working with Large Files (Chunking)

```python
chunk_size = 1024                          # set the chunk size
with open("large_file.txt", "rb") as file: # open the file in read mode
    while True:                            # loop through the file
        chunk = file.read(chunk_size)      # read the file
        if not chunk:                      # if the file is empty
            break                          # break the loop
        print(chunk)                       # print the chunk
        # Process chunk
```
